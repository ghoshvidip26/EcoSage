```python
from langchain_ollama import ChatOllama 
```

```python
model=ChatOllama(model="llama3.1")
```

```python
from pydantic import BaseModel,Field
```

```python
class EcoAdvice(BaseModel):
    # N	P	K	temperature	humidity
    N: str = Field(..., description="Nitrogen")
    P: str = Field(..., description="Phosphorus")
    K: str = Field(..., description="Potassium")
    temp: str = Field(..., description="Temperature")
    rainfall: str = Field(..., description="Rainfall")
```

```python
user_input = "Suggest the best crop for N=60, P=55, K=50, temperature=28°C, and rainfall=180mm."
```

```python
system_prompt = """ You are a Q&A assistant. Your goal is to help farmers with the crop recommendation if the query includes N,P,K,temperature,rainfall. """
try: 
    response = model.invoke(f"{system_prompt}\nUser: {user_input}")
    print(response)
except Exception as e:
        print(f"⚠️ Router LLM failed: {e}")
```
